**Data Analysis Questions**

```python
# What is the average score in math (G3)?

average_G3=Student['G3'].mean()
average_G3
```

```python
# How many students scored above 15 in their final grade (G3)?

Student[Student['G3']>15].sum().shape[0]
```

```python
# Is there a correlation between study time (study time) and the final grade (G3)?

Student['studytime'].corr(Student['G3'])
```

```python
# Which gender has a higher average final grade (G3)?

avg_by_gender=Student.groupby('sex')['G3'].mean()
print ("Average G3 by gender",avg_by_gender)

if avg_by_gender['F']>avg_by_gender['M']:
  print("Females has higher average in final grade G3")
else:
  print("Males has higher average in final grade G3")
```

**Data Visualisation**

```python
# Plot a histogram of final grades (G3).

plt.figure(figsize=(6, 4))
plt.hist(Student['G3'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Final Grades (G3)', fontsize=16)
plt.xlabel('Final Grade (G3)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a scatter plot between study time (study time) and final grade (G3).

plt.figure(figsize=(6,4))
plt.scatter(Student['studytime'], Student['G3'], color='blue', alpha=0.6, edgecolor='black')
plt.title('Scatter Plot: Study Time vs Final Grade', fontsize=16)
plt.xlabel('Study Time', fontsize=14)
plt.ylabel('Final Grade (G3)', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Create a bar chart comparing the average scores of male and female students.

plt.figure(figsize=(8, 6))
avg_by_gender.plot(kind='bar', color=['blue', 'red'], edgecolor='black', alpha=0.5)
plt.title('Average Final Grades by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Average Final Grade (G3)', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
